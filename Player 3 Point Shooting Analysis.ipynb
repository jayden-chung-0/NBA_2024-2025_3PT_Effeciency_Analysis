```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('NBA Stats 202425 All Metrics  NBA Player Props Tool.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
above_5_3PA_per_Game = df[(df['3PA'] / df['GP'] > 5) & (df['GP'] > 6)].copy()
```

```python
above_5_3PA_per_Game['3P_per_MIN'] = (above_5_3PA_per_Game['3PA'] * above_5_3PA_per_Game['3P%']) / (above_5_3PA_per_Game['MPG'] * above_5_3PA_per_Game['GP'])
```

```python
above_5_3PA_per_Game.sort_values(by = '3P_per_MIN', ascending = False, inplace = True)
```

```python
display(above_5_3PA_per_Game)
```

```python
above_5_3PA_per_Game.to_csv('3P per Min for all players above 5 3PA per GP and GP > 6.csv', index = False)
```
